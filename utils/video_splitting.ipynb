```python
import numpy as np
import cv2
import os
import time
import matplotlib.pyplot as plt
```

# Load video and split into frames

```python
video_main = os.path.normpath(os.getcwd() + (os.sep + os.pardir)*3 + os.sep + 'data' + os.sep + 'UCF-101')

with open(os.path.join(video_main,'classes.txt'),'r') as file:
    video_folders = [os.path.join(video_main,line[:-1]) for line in file]
#print(video_folders)

all_videos = [os.path.join(x,y) for x in video_folders for y in os.listdir(x)]
print(len(all_videos))
```

```python
total = []
for video_path in all_videos:
    vidcap = cv2.VideoCapture(video_path)
    total.append(int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT)))
    
#print(total)
```

```python
t_start = time.time()

vidcap = cv2.VideoCapture(all_videos[536])
frames = []
idx = [17,20,23,26,29,32]
for k in idx:
    vidcap.set(cv2.CAP_PROP_POS_FRAMES,k)
    success,image = vidcap.read()
    frames.append(image)
    print(image.shape)
    #cv2.imshow("Input", image)
    #cv2.WaitKey()
    
t_stop = time.time()
print(t_stop-t_start)

for k in range(len(frames)):
    cv2.imwrite('frame%d.png' % (k+1), frames[k])  
```
